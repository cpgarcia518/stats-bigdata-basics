## Conditionals (if / elif / else)

In Python, **conditionals** let your program make decisions.
They run different code depending on whether a condition is True or False.

**How it works**:
- `if` checks the first condition.
- `elif` (“else if”) checks another condition only if the previous one was False.
- `else` runs when none of the conditions were True.

````python
a = 5

if a > 0:
    print("a is positive")
elif a < 0:
    print("a is negative")
else:
    print("a is zero")
````

## Functions

A **function** is a reusable block of code that performs a task.

Instead of writing the same logic again and again, you give it a name and “call” it whenever you need it.

Think of it like a mini machine:
- you give it some inputs (called arguments)
- it does something
- it gives you back an output (using return)

````python
# Function definition
def myadd(a, b):
    return a + b # The return statement outputs the result

print("Pass arguments in order: ", myadd(5, 3))
print("Pass arguments by names: ", myadd(a=5, b=3))
print("Pass arguments by names: ", myadd(b=3, a=5))
````

## When should you write a function?

* When you notice you’re repeating the same code in more than one place.

* When you want to give a clear name to a task, so your code becomes easier to read and reuse.

* When you want to generalize an idea (same logic, different inputs), like adding any two numbers—not just `5 + 3`.

## Iterations

An **iteration** means repeating an action multiple times—usually to process several items in a collection like a list.

> Without iteration: printing items one by one

````python
this_is_a_list = [1, 2, "Hello"]

print(this_is_a_list[0])
print(this_is_a_list[1])
print(this_is_a_list[2])
````

This works… but it has two problems:
- It’s manual (you have to write one print() per element)
- It doesn’t scale: if the list has 100 items, you don’t want 100 print lines

With iteration: using a `for` **loop**

A **for loop** lets Python go through the list *automatically*, one element at a time:

````python
this_is_a_list = [1, 2, "Hello"]

for element in this_is_a_list:
    print(element)
````

## The Data

Data analysis starts **before** any model or statistical test.
If the data are messy or misunderstood, the results will be misleading.

### Common issues with data

#### 1) Type

Data can come in many forms:

* **Quantitative** (numbers)
* **Qualitative** (categories, text)
* **Structured / semi-structured / unstructured** (tables vs. logs vs. images, etc.)

#### 2) Quality

Data are almost never perfect. Typical problems include:

* **Missing values** (empty cells, “NA”, nulls)
* **Inconsistencies** (same thing written in different ways: `"M"`, `"Male"`, `"male"`)
* **Outliers** (values very different from the rest, sometimes real, sometimes errors)

A key habit: always ask **how the data were collected** (or how you should collect them).
Bad collection → biased results.

### Garbage in, garbage out

Better data quality → better results. <span style="color:red">garbage-in-garbage-out</span>

That’s why we often do **pre-processing**:

* Transform and clean the data **before** analysis to make the next steps easier and more reliable.

## Data Types (Variables)

In statistics, variables are often grouped into two big families:

### Categorical (Discrete)

* **Nominal:** categories with **no natural order**
  Examples: sex, colors, names, country
* **Ordinal:** categories with a **meaningful order**
  Examples: severity levels (low/medium/high), education level, satisfaction scores

### Numerical (Continuous)

* **Interval:** numeric values where **0 is arbitrary**
  Example: temperature in Celsius (0°C doesn’t mean “no temperature”)
* **Ratio:** numeric values with a **true zero**
  Examples: length, mass, counts, temperature in Kelvin

## Data Distributions

A **distribution** tells us:

* which values a variable can take, and
* how often each value appears.

One of the most important distributions in statistics is the **Gaussian** (also called the **Normal**) distribution:

![](images/normal_distribution.png)

### Why is the Normal distribution so important?

Because it shows up everywhere—especially when a value is the result of **many small independent effects**.

---

## Central Limit Theorem (CLT)

The **Central Limit Theorem** explains why the Normal distribution appears so often.

If you take many **independent** random variables (x_i) (they don’t influence each other), and add them:

[
y = \sum_{i=1}^{n} x_i
]

then for large (n), the sum (y) tends to follow a **Normal (Gaussian)** distribution—even if the individual (x_i) are not Normal.

This is one reason why the Normal distribution is used so much in real-world data analysis.

## Descriptive Statistics

Descriptive statistics, as for the name, are used to describe data distributions. The most common are:

- **Mean**: the “average” value (sum of all values divided by how many values you have)

$$\mu_x = \frac{1}{N}\sum_{i=1}^{N} x_i$$

- **Median**: the middle value when the data are sorted (50% of the values are below it, 50% are above it)
- **Percentile (or Quantile)**: the value below which a given percentage of the data falls.
- **Variance**: measures how spread out the data are around the mean (average squared distance from the mean)

$$\sigma^2 = \frac{1}{N-1}\sum_{i=1}^{N}(x_i - \mu_x)^2$$

- **Standard Deviation**: the square root of the variance (same “unit” as the original data), often written as $\sigma$

$$\sigma = \sqrt{\sigma^2} = \sqrt{\frac{1}{N-1}\sum_{i=1}^{N}(x_i - \mu_x)^2}$$

````python
# Descriptive statistics for Total Milk Yield (tmy)
tmy = cows["tmy"]  # pandas will ignore missing values in most summary methods

print("Mean (average):", round(tmy.mean(), 2))
print("Median (50th percentile):", round(tmy.median(), 2))
print("Std Dev (spread):", round(tmy.std(), 2))
print("75th percentile:", round(tmy.quantile(0.75), 2))
print("Variance:", round(tmy.var(), 2))

# Full summary table (count, mean, std, min, quartiles, max)
display(tmy.describe())
````

A **histogram** can show these descriptive statistics visually: where most values are, how spread out they are, and where the **mean**, **median**, and **±1 standard deviation** sit.


````python
# Visualizing mean, median, and standard deviation on a histogram
tmy = cows["tmy"].dropna()

mu = tmy.mean()
med = tmy.median()
sigma = tmy.std()

fig, ax = plt.subplots(figsize=(8, 4))

sns.histplot(tmy, ax=ax, kde=True, bins=20)

# Mean and median
ax.axvline(mu, lw=2, color="k", ls="--", label="Mean")
ax.axvline(med, lw=2, color="r", ls="--", label="Median")

# One standard deviation from the mean
ax.axvline(mu + sigma, lw=2, ls="-.", label=r"$\mu \pm \sigma$")
ax.axvline(mu - sigma, lw=2, ls="-.")

ax.set(xlabel="Total Milk Yield (tmy)", ylabel="Count", title="Distribution of tmy")
ax.legend()

plt.tight_layout()
plt.show()
````

## Correlations

A **correlation** measures how two variables **move together**.

It answers questions like:

* When **x** increases, does **y** also tend to increase?
* When **x** increases, does **y** tend to decrease?
* Or do they seem unrelated?

A correlation coefficient usually lies in the range **([-1, 1])**:

* **+1** → perfect positive association (they increase together)
* **-1** → perfect negative association (one increases while the other decreases)
* **0** → no *linear* association (for Pearson)

> Important: “0 correlation” does **not** always mean “no relationship” — the relationship could be non-linear.

---

### Pearson correlation (linear association)

**Pearson** correlation measures **linear** association between two numerical variables.

$$\rho = \frac{cov(x, y)}{\sigma_x \sigma_y}$$

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/3/34/Correlation_coefficient.png/400px-Correlation_coefficient.png" width="450"/>

Source: [Wikipedia](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient)

---

### Spearman correlation (rank / monotonic association)

**Spearman** correlation is Pearson correlation applied to the **ranks** of the data.
It can capture **monotonic** relationships, even when they are not perfectly linear.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/4/4e/Spearman_fig1.svg/300px-Spearman_fig1.svg.png" width="420"/>

Source: [Wikipedia](https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient)

---

### Correlation does not imply causation

Two variables can be strongly correlated even if:

* neither causes the other, or
* both are influenced by a third factor, or
* it’s just a coincidence in the data.

A fun reminder: [Spurious Correlations](https://www.tylervigen.com/spurious-correlations)


````python
# Correlation between Milk Yield (tmy) and Milking Interval (mi)

# Select the two columns and remove rows with missing values
df_corr = cows[["tmy", "mi"]].dropna()

print("Correlation between Total Milk Yield (tmy) and Milking Interval (mi)")

# Pearson correlation (linear relationship)
pearson_corr = df_corr["tmy"].corr(df_corr["mi"], method="pearson")
print("Pearson (linear):", round(pearson_corr, 2))

# Spearman correlation (rank / monotonic relationship)
spearman_corr = df_corr["tmy"].corr(df_corr["mi"], method="spearman")
print("Spearman (rank):", round(spearman_corr, 2))
````

$$E(w) = \sum_{i=0}^{M} (\hat y_i - y_i)^2$$

All the minimization algorithms are already implemented in many libraries.
This technique is called **Ordinary Least Square** (OLS).

## Linear Models

What is a linear regression?

Mathematically, the general shape of a linear model is:

$$\hat y = w_0 + \sum_{i=1}^{N} w_i x_i = w_0 + w_1 x_1 + w_2 x_2 + ... + w_N x_N$$

Where:

* $w_0$ is the **intercept** (the value of $\hat y$ when all $x_i = 0$)
* $w_i$ are the **coefficients** (how much $\hat y$ changes when $x_i$ increases by 1 unit, keeping the others fixed)
* $\hat y$ is the **predicted** value

The objective is to find the best set of weights $\bar{w}$ that minimizes the error between the predicted values $\hat y$ and the true values $y$.

A common choice is the **sum of squared errors**:

$$E(w) = \sum_{j=1}^{M} (\hat y_j - y_j)^2$$

All the minimization algorithms are already implemented in many libraries.
This technique is called **Ordinary Least Squares** (OLS).

## Statsmodels

`statsmodels` is one of the main Python libraries for **statistical modeling**.
It is especially useful when you want:

* classical statistical models (like linear regression)
* detailed output (coefficients, standard errors, p-values, confidence intervals)
* results presented in a style similar to what you may see in **R**

---

### A first linear model

The first model is a simple regression where **Height** depends on **Weight**:

$$H \sim W$$

This notation means: “model (H) using (W) as a predictor”.

In equation form:

$$H = w_0 + w_1 W$$

Where:

* $w_0$ is the **intercept** (also called **bias**): the predicted height when $W = 0$
* $w_1$ is the **slope** (coefficient): how much $H$ changes when $W$ increases by 1 unit


````python
import statsmodels.formula.api as smf

# Define the model using a formula
m1 = smf.ols(formula="height ~ 1 + weight", data=df)

# Fit the model (estimate the coefficients)
res1 = m1.fit()

# Print a detailed summary of the results
print(res1.summary())
````

Given the results, our model is:

$$H = 113.88 + 0.905 \cdot W$$

## Interpretations

From the fitted model:

* The slope ($w_1$) tells us the expected change in height for a **1 kg** increase in weight.
  For example, if $w_1 \approx 0.90$, then **a person who is 1 kg heavier is expected (on average) to be about 0.90 cm taller**.

* The $R^2$ value describes how much of the variability in height is explained by the model.
  If $R^2 \approx 0.60$, it suggests that **weight explains about 60% of the variation in height** (in this dataset).

* The **95% confidence interval** for $w_1$ gives a plausible range for the true slope.
  For example, if it is ([0.82,\ 0.99]), it means the data support a positive relationship in that range.

Overall, this is strong evidence of an association between **height** ($H$) and **weight** ($W$).

### A note about the intercept

Intercepts are often tricky to interpret. Here, the intercept is the predicted height when $W = 0$ kg, which is not realistic for a person—so the intercept is mostly a mathematical part of the line, not a meaningful “real-world” value.

---

## P-values

A **p-value** answers this question:

> If the *null hypothesis* were true, how surprising would it be to see a result at least as extreme as the one we observed?

(Definition source: Wikipedia)

In this model, the null hypothesis is:

* $H_0$: weight has no effect on height
  (equivalently: the coefficient $w_1 = 0$)

If the p-value is less than a chosen threshold (often $T = 0.05$), it suggests the result would be unlikely under $H_0$, so we **reject** $H_0$ and say weight is **statistically significant**.

### A gentle warning

Significant does not automatically mean:

* “important in practice”
* “causal”
* “the model is correct”

So it’s always good to look at the plot, check assumptions, and think about whether the conclusion makes sense.

For more: [Misuse of p-values](https://en.wikipedia.org/wiki/Misuse_of_p-values)

> *All models are wrong, but some are useful.*
> — George Box (1976)


````python
# Plot data + fitted linear model
plot_df = df.assign(fitted=res1.fittedvalues).sort_values("weight")

ax = sns.scatterplot(data=df, x="weight", y="height", alpha=0.5, label="Data")
sns.lineplot(data=plot_df, x="weight", y="fitted", lw=2, label="Model", ax=ax)

ax.set(xlabel="Weight (kg)", ylabel="Height (cm)", title="Linear Model: Height vs Weight")
ax.legend()
plt.tight_layout()
plt.show()
````

## Polynomial Regression and Standardization

### Polynomial regression

A **polynomial regression** is still a *linear model in the coefficients*—but it uses **power terms** of the same variable as additional features.

For example, a 2nd-order (quadratic) model is:

$$H = w_0 + w_1 W + w_2 W^2$$

* If $w_2 = 0$, the model becomes a straight line.
* If $w_2 \neq 0$, the curve bends: the sign and size of $w_2$ control the **curvature**.

So instead of fitting a straight line, the model can fit a **parabola**.

### Why standardize before adding powers?

When you add terms like $W$ and $W^2$, they can become highly related to each other. This can create:

* **multicollinearity** (predictors strongly correlated)
* **numerical issues** (very large values, unstable fits)

A common fix is to **standardize** the input feature first.

---

### Standardization

Standardization rescales a variable so that it has:

* mean (0)
* standard deviation (1)

$$\hat x = \frac{x - \mu_x}{\sigma_x}$$

Where:

* $\hat x$ is the **standardized** variable
* $x$ is the original variable
* $\mu_x$ is the mean of $x$
* $\sigma_x$ is the standard deviation of $x$

**Why it helps:**

* reduces multicollinearity when using polynomial terms
* improves numerical stability (especially with large/small scales)
* makes coefficients more comparable across predictors

**Trade-off:**

* coefficients are a bit less “direct” to interpret, because they refer to standardized units rather than the original units.


````python
# Standardize
howell["weight_s"] = (howell["weight"] - howell["weight"].mean()) / howell["weight"].std()

# Plot The difference between standardized and non-standardized weight
ax = sns.histplot(data=howell, x="weight", alpha=0.4, color="salmon", label="Non-Standardized", kde=True)
sns.histplot(data=howell, x="weight_s", alpha=0.4, color="lightblue", ax=ax, label="Standardized", kde=True)

ax.axvline(0, ls="--", color="k")
ax.set(xlabel="Weight (kg)", ylabel="Count", title="Weight Distribution")
ax.legend()

plt.tight_layout()
plt.show()
````

````python
# Fit the polynomial model:

m3 = smf.ols(formula="height ~ 1 + weight_s + I(weight_s**2)", data=howell)
res3 = m3.fit()
print(res3.summary())
````

## Multivariate Linear Models

### *Or simply Linear Models*

The concept is the same as for linear models, but instead of one predictor we use multiples!

In the precedent case we may add the categorical variables "male" to the model:

$$H = w_0 + w_1 W + w_2 M$$

The question we are asking now is: *does the model change if the subject is male?*



````python
# We go back to adult subjects

m5 = smf.ols(formula="height ~ 1 + weight + C(male)", data=df)
res5 = m5.fit()
print(res5.summary())
````

The new model for *male* subject is:

$$H = 122.70 + 0.64 \cdot W + 6.5 \cdot M$$

and the new model for *female* subject is:

$$H = 122.70 + 0.64 \cdot W$$

````python
ax = sns.scatterplot(data=df, x="weight", y="height", alpha=0.5, color="grey", hue="male", palette={1: "tab:blue", 0: "tab:orange"})

male_model = 122.7034 + df["weight"] * 0.6412 + 6.5003 * 1
sns.lineplot(x=df["weight"], y=male_model, color="blue", label="Male Model", ax=ax)

female_model = 122.7034 + df["weight"] * 0.6412 + 6.5003 * 0
sns.lineplot(x=df["weight"], y=female_model, color="r", label="Female Model", ax=ax)

original_model = 113 + df["weight"] * 0.90
sns.lineplot(x=df["weight"], y=original_model, color="k", label="Original Model", ax=ax, ls="--")

plt.tight_layout()
plt.show()
````
